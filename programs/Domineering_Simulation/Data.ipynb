```python
import matplotlib.pyplot as plt
from programs.Domineering_Simulation.Domineering_Simulation import *

player = 1
file = 'graphs.txt'
height = 5
length = 5
depth = 5
    
accuracy_list = []
precision_list = []
depth_list = []
for i in range(200):
    generate_graphs_txt(random.randint(5,9),[height,length])
    graph = generate_empty_graph(height=height, length=length)
    board = generate_empty_board(height, length)
    board = fill_board(file=file, board=board)
    build_graph_from_board(graph=graph, board=board)
    accuracy, precision = accuracy_comparison(graph=graph, board=board, player=player, depth=5)
    depth = depth_comparison(graph=graph, board=board, player=player)
    depth_list.append(depth)
    if accuracy and precision != -99:
        accuracy_list.append(accuracy)
        precision_list.append(precision)
```

```python
plt.hist(accuracy_list, 
         histtype='bar',
         facecolor='b',
         alpha=0.5)

plt.xlabel("Fraction of Overall Games won \n (1.0 = 100%)")
plt.ylabel("Times Occurred")
plt.title("Overall Possible Wins in Given Board")
plt.show()
```

```python
plt.hist(precision_list, 
         histtype='bar',
         facecolor='b',
         alpha=0.5)

plt.xlabel("Fraction of Correct Declaration \n (1.0 = 100%)")
plt.ylabel("Times Occurred")
plt.title("Games in Which Wincon Minimax Correctly Declares a Victory")
plt.show()
```
